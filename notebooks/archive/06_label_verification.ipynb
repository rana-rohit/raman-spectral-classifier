```python
import numpy as np
import pandas as pd

from metadata.label_metadata import (
    LABEL_METADATA,
    get_strain_name,
    get_antibiotic_name,
    get_biological_category,
    get_full_metadata
)
```

```python
DATA_DIR = "../data/raw"

X_REFERENCE_PATH = os.path.join(DATA_DIR, "X_reference.npy")
Y_REFERENCE_PATH = os.path.join(DATA_DIR, "y_reference.npy")

X_reference = np.load(X_REFERENCE_PATH)
y_reference = np.load(Y_REFERENCE_PATH)

print("X_reference shape:", X_reference.shape)
print("y_reference shape:", y_reference.shape)
```

```python
unique_labels = np.unique(y_reference)

print(unique_labels)
```

```python
rows = []

for label in unique_labels:

    rows.append({
        "Label ID": label,
        "Strain": get_strain_name(label),
        "Antibiotic": get_antibiotic_name(label),
        "Category": get_biological_category(label)
    })

label_df = pd.DataFrame(rows)

label_df
```

```python
distribution_rows = []

for label in unique_labels:

    count = np.sum(y_reference == label)

    distribution_rows.append({
        "Label ID": label,
        "Strain": get_strain_name(label),
        "Samples": count,
        "Antibiotic": get_antibiotic_name(label),
        "Category": get_biological_category(label)
    })

distribution_df = pd.DataFrame(distribution_rows)

distribution_df.sort_values("Label ID")
```

```python
Y_2018_PATH = os.path.join(DATA_DIR, "y_2018clinical.npy")
Y_2019_PATH = os.path.join(DATA_DIR, "y_2019clinical.npy")

y_2018 = np.load(Y_2018_PATH)
y_2019 = np.load(Y_2019_PATH)

print("2018 Clinical Labels:")
print(np.unique(y_2018))

print("\n2019 Clinical Labels:")
print(np.unique(y_2019))
```

```python
print("2018 Clinical Mapping")
print("-" * 50)

for label in np.unique(y_2018):

    print(
        f"{int(label)} -> "
        f"{get_strain_name(int(label))}"
    )

print("\n2019 Clinical Mapping")
print("-" * 50)

for label in np.unique(y_2019):

    print(
        f"{int(label)} -> "
        f"{get_strain_name(int(label))}"
    )
```
